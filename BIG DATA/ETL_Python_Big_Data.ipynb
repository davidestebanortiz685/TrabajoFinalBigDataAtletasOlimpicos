```python
import pandas as pd
import sqlite3
import numpy as np

print("Librerías cargadas correctamente ✅")
```

```python
# Cargar archivos directamente desde Colab
athletes = pd.read_csv('Athletes_Games.csv')
edition  = pd.read_csv('Edition_Games.csv')
equipos  = pd.read_csv('equipos_Games.csv')
events   = pd.read_csv('Competition_Games.csv')
jjoo     = pd.read_csv('JJ.OO.csv')

print("Athletes:", athletes.shape)
print("Edition:", edition.shape)
print("Equipos:", equipos.shape)
print("Events:", events.shape)
print("JJOO:", jjoo.shape)
```

```python
print("ATHLETES ")
print(athletes.head(3))
print("\nEDITION")
print(edition.head(3))
print("\nEQUIPOS ")
print(equipos.head(3))
print("\n EVENTS")
print(events.head(3))
print("\n JJOO ")
print(jjoo.head(3))
```

```python
# Verificar nulos en cada tabla
print("=== NULOS POR TABLA ===")
print("\nAthletes:")
print(athletes.isnull().sum())
print("\nEdition:")
print(edition.isnull().sum())
print("\nEquipos:")
print(equipos.isnull().sum())
print("\nEvents:")
print(events.isnull().sum())
print("\nJJOO:")
print(jjoo.isnull().sum())
```

```python
# Limpiar athletes - convertir tipos y manejar nulos
athletes['Age']    = pd.to_numeric(athletes['Age'],    errors='coerce')
athletes['Height'] = pd.to_numeric(athletes['Height'], errors='coerce')
athletes['Weight'] = pd.to_numeric(athletes['Weight'], errors='coerce')

# Limpiar JJOO - estandarizar medallas
jjoo['Medal'] = jjoo['Medal'].fillna('No Medal')

print("Nulos en Age:",    athletes['Age'].isnull().sum())
print("Nulos en Height:", athletes['Height'].isnull().sum())
print("Nulos en Weight:", athletes['Weight'].isnull().sum())
print("Limpieza completada ✅")
```

```python
# Indicador 1 - Total participaciones
total_participaciones = len(jjoo)

# Indicador 2 - Total medallas
total_medallas = len(jjoo[jjoo['Medal'] != 'No Medal'])

# Indicador 3 - Porcentaje de medallistas
pct_medallistas = (total_medallas / total_participaciones) * 100

# Indicador 4 - Promedio edad, altura, peso
avg_edad   = athletes['Age'].mean()
avg_altura = athletes['Height'].mean()
avg_peso   = athletes['Weight'].mean()

print(f"Total participaciones: {total_participaciones:,}")
print(f"Total medallas:        {total_medallas:,}")
print(f"% medallistas:         {pct_medallistas:.2f}%")
print(f"Promedio edad:         {avg_edad:.1f} años")
print(f"Promedio altura:       {avg_altura:.1f} cm")
print(f"Promedio peso:         {avg_peso:.1f} kg")
```

```python
# Crear base de datos SQLite
conn = sqlite3.connect('JuegosOlimpicos.db')

# Guardar cada tabla
athletes.to_sql('dim_atleta',        conn, if_exists='replace', index=False)
edition.to_sql('dim_edicion',        conn, if_exists='replace', index=False)
equipos.to_sql('dim_equipo',         conn, if_exists='replace', index=False)
events.to_sql('dim_evento',          conn, if_exists='replace', index=False)
jjoo.to_sql('fact_participacion',    conn, if_exists='replace', index=False)

print("Base de datos SQLite creada ✅")

# Verificar
for tabla in ['dim_atleta','dim_edicion','dim_equipo','dim_evento','fact_participacion']:
    count = pd.read_sql(f"SELECT COUNT(*) as filas FROM {tabla}", conn)
    print(f"{tabla}: {count['filas'][0]:,} filas")

conn.close()
```

#DASHBOARD PYTHON





```python
import plotly.express as px
import plotly.graph_objects as go

# Reconectar a SQLite
conn = sqlite3.connect('JuegosOlimpicos.db')

print("Plotly listo ✅")
```

```python
# Consulta
df_paises = pd.read_sql("""
    SELECT e.NOC, COUNT(*) as participaciones
    FROM fact_participacion f
    JOIN dim_equipo e ON f.team_id = e.team_id
    GROUP BY e.NOC
    ORDER BY participaciones DESC
    LIMIT 10
""", conn)

# Gráfico
fig1 = px.bar(df_paises,
              x='participaciones',
              y='NOC',
              orientation='h',
              title='Top 10 Países por Participación Histórica',
              color='participaciones',
              color_continuous_scale='Blues')
fig1.update_layout(yaxis={'categoryorder':'total ascending'})
fig1.show()
```

```python
df_años = pd.read_sql("""
    SELECT e.Year, COUNT(*) as participaciones
    FROM fact_participacion f
    JOIN dim_edicion e ON f.edition_id = e.edition_id
    GROUP BY e.Year
    ORDER BY e.Year
""", conn)

fig2 = px.line(df_años,
               x='Year',
               y='participaciones',
               title='Evolución Histórica de Participaciones 1896-2016',
               markers=True)
fig2.update_traces(line_color='#1f77b4')
fig2.show()
```

```python
df_sexo = pd.read_sql("""
    SELECT a.Sex, COUNT(*) as cantidad
    FROM fact_participacion f
    JOIN dim_atleta a ON f.athlete_id = a.athlete_id
    GROUP BY a.Sex
""", conn)

fig3 = px.pie(df_sexo,
              values='cantidad',
              names='Sex',
              title='Distribución de Participación por Género',
              hole=0.4,
              color_discrete_map={'M':'#1f77b4','F':'#ff7f0e'})
fig3.show()
```

### Visualización: Top 10 países con más medallas de oro,plata,bronce históricamente

```python
conn = sqlite3.connect('JuegosOlimpicos.db') # Reopen the connection
df_gold_medals = pd.read_sql("""
    SELECT te.NOC, COUNT(*) as total_medallas
    FROM fact_participacion fp
    JOIN dim_equipo te ON fp.team_id = te.team_id
    WHERE fp.Medal IN ('Gold','Bronze','Silver')
    GROUP BY te.NOC
    ORDER BY total_medallas DESC
    LIMIT 10
""", conn)

fig_gold = px.bar(df_gold_medals,
                  x='total_medallas',
                  y='NOC',
                  orientation='h',
                  title='Top 10 Países con más Medallas (Oro, Plata, Bronce) Históricamente',
                  color='total_medallas',
                  color_continuous_scale='YlGnBu')
fig_gold.update_layout(yaxis={'categoryorder':'total ascending'})
fig_gold.show()
```

### Visualización: Proporción de Oro, Plata y Bronce del total de medallas

```python
df_medal_distribution = pd.read_sql("""
    SELECT Medal, COUNT(*) as cantidad
    FROM fact_participacion
    WHERE Medal != 'No Medal'
    GROUP BY Medal
""", conn)

fig_medal_pie = px.pie(df_medal_distribution,
                       values='cantidad',
                       names='Medal',
                       title='Distribución de Medallas (Oro, Plata, Bronce)',
                       hole=0.4,
                       color_discrete_map={
                           'Gold':'#FFD700',
                           'Silver':'#C0C0C0',
                           'Bronze':'#CD7F32'
                       })
fig_medal_pie.show()
```
