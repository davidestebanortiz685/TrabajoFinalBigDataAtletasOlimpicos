```python
import unicodedata
import re

def limpiar_dataframe(df):
    df_clean = df.copy()

    def limpiar_columna(col):
        # Quitar espacios extremos
        col = col.strip()
        # Convertir a minúsculas
        col = col.lower()
        # Quitar tildes
        col = unicodedata.normalize('NFKD', col).encode('ascii', 'ignore').decode('utf-8')
        # Reemplazar espacios por _
        col = col.replace(' ', '_')
        # Eliminar caracteres especiales (solo letras, números y _)
        col = re.sub(r'[^a-z0-9_]', '', col)
        # Reemplazar múltiples _ por uno solo
        col = re.sub(r'_+', '_', col)
        return col

    df_clean.columns = [limpiar_columna(col) for col in df_clean.columns]

    return df_clean
```
