```python
# Instalar librerías
!pip install mlxtend

# Importar librerías
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
file_path = 'movies_transactions_apriori.csv'
df = pd.read_csv(file_path)
```

```python
transactions = []

for _, row in df.iterrows():
    items = [str(x) for x in row.dropna().tolist()]
    transactions.append(items)

# Transformar datos al formato binario
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
movies_df = pd.DataFrame(te_array, columns=te.columns_)

# Aplicar Apriori
frequent_itemsets = apriori(
    movies_df,
    min_support=0.10,
    use_colnames=True
)

# Generar reglas de asociación
rules = association_rules(
    frequent_itemsets,
    metric='confidence',
    min_threshold=0.60
)

# Filtrar por lift mínimo
rules = rules[rules['lift'] >= 1.2]

# Ordenar reglas
rules = rules.sort_values(by=['lift', 'confidence'], ascending=False)

# Mostrar resultados
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

# 3. Parámetros utilizados

Se utilizaron los siguientes valores:

| Parámetro      | Valor |
| -------------- | ----- |
| min_support    | 0.10  |
| min_confidence | 0.60  |
| min_lift       | 1.20  |

## Explicación

* **Support (Soporte):** frecuencia con la que aparece una combinación de películas.
* **Confidence (Confianza):** probabilidad de que se vea una película cuando otra ya fue vista.
* **Lift:** mide la fuerza real de la asociación. Un lift mayor a 1 indica relación positiva.
